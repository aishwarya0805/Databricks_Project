Total appointments and no-shows

```python
%sql
SELECT 
    COUNT(*) AS TotalAppointments,
    SUM(CASE WHEN CleanedStatus = 'did not attend' THEN 1 ELSE 0 END) AS NoShows,
    ROUND(
        SUM(CASE WHEN CleanedStatus = 'did not attend' THEN 1 ELSE 0 END) 
        / COUNT(*), 
        2
    ) AS NoShowRate
FROM appointments;
```

No-show by Gender

```python
%sql
SELECT Gender,
    COUNT(*) AS TotalAppointments,
    SUM(CASE WHEN CleanedStatus = 'did not attend' THEN 1 ELSE 0 END) AS NoShows,
    ROUND(
        SUM(CASE WHEN CleanedStatus = 'did not attend' THEN 1 ELSE 0 END) 
        / COUNT(*), 
        2
    ) AS NoShowRate
FROM appointments a join patients p on a.PatientID = p.PatientID
GROUP BY Gender;
```

No-show by Weekday

```python
%sql
SELECT AppointmentDay,
    COUNT(*) AS TotalAppointments,
    SUM(CASE WHEN CleanedStatus = 'did not attend' THEN 1 ELSE 0 END) AS NoShows,
    ROUND(
        SUM(CASE WHEN CleanedStatus = 'did not attend' THEN 1 ELSE 0 END) 
        / COUNT(*), 
        2
    ) AS NoShowRate
FROM appointments
GROUP BY AppointmentDay
ORDER BY noshows DESC,AppointmentDay ASC;
```

No-show by Age Group

```python
%sql
SELECT CASE 
         WHEN p.Age BETWEEN 0 AND 18 THEN 'Child'
         WHEN p.Age BETWEEN 19 AND 35 THEN 'Young Adult'
         WHEN p.Age BETWEEN 36 AND 60 THEN 'Adult'
         ELSE 'Senior'
       END AS AgeGroup,
    COUNT(*) AS TotalAppointments,
    SUM(CASE WHEN CleanedStatus = 'did not attend' THEN 1 ELSE 0 END) AS NoShows,
    ROUND(
        SUM(CASE WHEN CleanedStatus = 'did not attend' THEN 1 ELSE 0 END) 
        / COUNT(*), 
        2
    ) AS NoShowRate
FROM appointments a JOIN patients p ON a.PatientID = p.PatientID
GROUP BY agegroup
ORDER BY noshows DESC;
```

Waiting Time Impact

```python
%sql
SELECT datediff(AppointmentDate, ScheduledDate) AS WaitingDays,
    COUNT(*) AS TotalAppointments,
    SUM(CASE WHEN CleanedStatus = 'did not attend' THEN 1 ELSE 0 END) AS NoShows,
    ROUND(
        SUM(CASE WHEN CleanedStatus = 'did not attend' THEN 1 ELSE 0 END) 
        / COUNT(*), 
        2
    ) AS NoShowRate
FROM appointments a JOIN patients p ON a.PatientID = p.PatientID
GROUP BY WaitingDays
HAVING WaitingDays >= 0
ORDER BY WaitingDays DESC;
```
